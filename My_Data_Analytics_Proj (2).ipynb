```python
df1=pd.read_csv(r"DuelNewSalary.csv")
df1['Total_sales'] = df1["January"]+df1["February"]+df1["March"]+df1["April"]+df1["May"]+df1["June"]
+df1["July"]+df1["August"]+df1["September"]+df1["October"]+df1["November"]+df1["December"]
df1
```

```python
df1['Comission']=df1['Total_sales']*0.05
df1['Salary']=df1['Comission']+3000
df1
```

```python

def increase(x):
    if x.Comission>4000:
        return x.Salary+x.Salary*0.2
    else:
        return x.Salary+x.Salary*0.1
df1["NewSalary"]=df1.apply(increase,axis=1)
df1
```

```python
splittedcolumns=df["Last,First"].str.split(',',expand=True)
df1["FirstName"]=splittedcolumns[0]
df1["LastName"]=splittedcolumns[1]
```

```python
color=pd.read_csv("Color.csv")
color.head(2)
```

```python
def bgc(s):
    if s["Product_Name"] == "Laptop":
        return["background-color:tan"]*31
    if s["Product_Name"] == "Camera":
        return["background-color:gold"]*31
    if s["Product_Name"] == "Computer":
        return["background-color:pink"]*31
    if s["Product_Name"] == "Mobile":
        return["background-color:silver"]*31
    else:
        return["background-color:red"]*31
color.style.apply(bgc,axis=1)
```

```python
import plotly.express as px

# Scatter plot: Age vs. Spending Score
fig1 = px.scatter(df1, x='Age', y='Spending Score (1-100)', color='Genre',
                  title='Age vs. Spending Score',
                  labels={'Age': 'Age', 'Spending Score (1-100)': 'Spending Score'},
                  hover_data=['Annual Income (k$)'])

fig1.show()
```

```python
# Bar plot: Total Sales by Genre
total_sales_by_genre = df1.groupby('Genre')['Total_sales'].sum().reset_index()

fig2 = px.bar(total_sales_by_genre, x='Genre', y='Total_sales',
              title='Total Sales by Genre',
              labels={'Genre': 'Gender', 'Total_sales': 'Total Sales'})
fig2.show()
```

```python
# Violin plot: Spending Score by Age Group
df1['Age_Group'] = pd.cut(df1['Age'], bins=[0, 20, 30, 40, 50, 60, 70, 80, 90, 100],
                          labels=['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100'])

fig = px.violin(df1, x='Age_Group', y='Spending Score (1-100)', box=True, points="all",
                  title='Spending Score by Age Group',
                  labels={'Age_Group': 'Age Group', 'Spending Score (1-100)': 'Spending Score'})

fig.show()
```

```python
import plotly.express as px
fig = px.histogram(df1, x="Age", y="Annual Income (k$)", color="Genre",
                   hover_data=df1)
fig.show()
```

```python
# 3D Scatter Plot: Age, Annual Income, and Spending Score
fig = px.scatter_3d(df1, x='Age', y='Annual Income (k$)', z='Spending Score (1-100)',
                     color='Genre', title='3D Scatter Plot: Age, Annual Income, and Spending Score',
                     labels={'Age': 'Age', 'Annual Income (k$)': 'Annual Income (k$)', 'Spending Score (1-100)': 'Spending Score'})

fig.show()
```

```python
import plotly.figure_factory as ff

# Correlation matrix
corr_matrix = df1[['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Total_sales']].corr()

fig = ff.create_annotated_heatmap(
    z=corr_matrix.values,
    x=list(corr_matrix.columns),
    y=list(corr_matrix.index),
    annotation_text=corr_matrix.round(2).values,
    colorscale='Viridis'
)

fig.update_layout(title='Correlation Matrix',
                   xaxis=dict(title='Features'),
                   yaxis=dict(title='Features'))

fig.show()
```

```python
# Pie chart: Distribution of Product Sales
product_sales = df1.groupby('Product Name')['Total_sales'].sum().reset_index()

fig = px.pie(product_sales, values='Total_sales', names='Product Name',
              title='Distribution of Product Sales')

fig.show()
```

```python
# Line plot: Monthly Sales Over Time
monthly_sales = df1[['January', 'February', 'March', 'April', 'May', 'June',
                     'July', 'August', 'September', 'October', 'November', 'December']].sum().reset_index()
monthly_sales.columns = ['Month', 'Total_Sales']

fig = px.line(monthly_sales, x='Month', y='Total_Sales',
               title='Monthly Sales Over Time',
               labels={'Month': 'Month', 'Total_Sales': 'Total Sales'})

fig.show()
```

```python
# Box plot: Spending Score by Genre
fig = px.box(df1, x='Genre', y='Spending Score (1-100)',
              title='Spending Score by Genre',
              labels={'Genre': 'Gender', 'Spending Score (1-100)': 'Spending Score'})

fig.show()
```

```python
# Histogram: Distribution of Annual Income
fig = px.histogram(df1, x='Annual Income (k$)',
                    title='Distribution of Annual Income',
                    labels={'Annual Income (k$)': 'Annual Income (k$)'},
                    nbins=20)

fig.show()
```
